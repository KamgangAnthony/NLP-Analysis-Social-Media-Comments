```python
#  Sentiment Analysis comment using NLTK library
cnter = 0
cnter1 = 0
cnter2 = 0
def sentimentAnalysis(sentencesComments):
    global cnter
    global cnter1
    global cnter2
    cnterr1 = 0
    cnterr2 = 0
    posY = []
    negY = []
    sid = SentimentIntensityAnalyzer()
    sumPos = 0
    sumNeg = 0
    cntxx1 = 0
    cntxx2 = 0
    cntxxx1 = []
    cntxxx2 = []
    old1 = 0
    old2 = 0
    for sentence in sentencesComments:
        ss = sid.polarity_scores(sentence)
        for type in sorted(ss):
            cnter += 1
            if (type == "neg" or type == "neu"):
                #sumNeg += ss[type]
                negY.append(ss[type])
                cntxx1 += 1
                cntxxx1.append(cntxx1)
                #cntxxx2.append(cntxx2)
                #posY.append(old2)
                #old1 = ss[type]
                #cntxx1 += 1
                cnter1 += 1
                cnterr1 += 1

            if (type == "pos"):
                #sumPos += ss[type]
                posY.append(ss[type])
                cntxx2+=1
                cntxxx2.append(cntxx2)
                #cntxxx1.append(cntxx1)
                #negY.append(old1)
                #old2 = ss[type]
                #cntxx2+=1
                cnter2 += 1
                cnterr2 += 1
    if (cnterr1 + cnterr2) != 0:
        pctneg = round(((cnterr1 / (cnterr1 + cnterr2)) * 100) , 2)
        pctpos = round(((cnterr2 / (cnterr1 + cnterr2)) * 100) , 2)
    if (cnterr1 + cnterr2) == 0:
        pctneg = 0
        pctpos = 0
    return posY, negY, cntxxx1, cntxxx2, pctneg, pctpos
```

```python
# For each Id_post, we run analysis and create graph temporal sentiment analysis of comments facebook
fi = [plt.figure()] * len(post_ids)
fi2 = [plt.figure()] * len(post_ids)
cnt = 0
for post_id in post_ids:
    for post in get_posts(post_urls=[post_id], credentials=('youremail', 'yourpassword'), options={"comments": True}):
        # Get created time of post
        created_time_post = post['time']
        # Get comment of post
        sentencesComments = []
        timeComments = []
        cntx = []
        cnt1 = 0
        for x in post['comments_full']:
            cnt1 = cnt1 + 1
            sentencesComments.append(x['comment_text'])
            timeComments.append(x['comment_time'])
            cntx.append(cnt1)
        # Convert time data
        #timeX = cntx
        # Convert sentiment analysis data
        posY, negY, timeX1, timeX2, pctneg, pctpos = sentimentAnalysis(sentencesComments)

        plt.scatter(timeX2, posY) #  it will show exactly points if you need
        plt.scatter(timeX1, negY)

        fi[cnt] = plt.figure()
        #fig2 = plt.plot(timeX1, negY, color = 'r')
        #fig1 = plt.plot(timeX2, posY, color = 'g')

        #ax1 = fi[cnt].add_subplot(221)
        #ax2 = fi[cnt].add_subplot(222)
        #if len(timeX1) > len(timeX2):
        #    ax1.plot(timeX1, posY, color = 'g')
        #    ax2.plot(timeX1, negY, color = 'r')

        #if len(timeX2) > len(timeX1):
        #    ax1.plot(timeX2, posY, color = 'g')
        #    ax2.plot(timeX2, negY, color = 'r')

        #if len(timeX2) == len(timeX1):
        #    ax1.plot(timeX2, posY, color = 'g')
        #    ax2.plot(timeX2, negY, color = 'r')

        plt.xlabel('Number of comments')
        plt.ylabel('Sentiment analysis')
        #fi[cnt].suptitle(post_titles[cnt])
        cnt += 1


# Show graph
show()
```
