# Function get all coments from id-post
def getComments(id_post):
    sentencesComments = []
    timeComments = []
    cnt = 0
    for post in get_posts(post_urls=id_post, credentials=('youremail', 'yourpassword'), options={"comments": True}):
        try:
            for x in post['comments_full']:
                cnt = cnt + 1
                print('Comment {0} : {1} -- Time = {2}'.format(cnt, x['comment_text'], x['comment_time']))
                sentencesComments.append(x['comment_text'])
                timeComments.append(x['comment_time'])

        except:
            continue
    return sentencesComments, timeComments

# Convert time comment to coordinate X in Graph
def timeToX(timeComments, created_time_post):
    timeX = []
    for timeComment in timeComments:
        distTime = timeComment - created_time_post
        x = int(distTime.total_seconds())
        timeX.append(x)
    return timeX

```python
#  Sentiment Analysis comment using NLTK library
cnter = 0
cnter1 = 0
cnter2 = 0
def sentimentAnalysis(sentencesComments):
    global cnter
    global cnter1
    global cnter2
    cnterr1 = 0
    cnterr2 = 0
    posY = []
    negY = []
    sid = SentimentIntensityAnalyzer()
    sumPos = 0
    sumNeg = 0
    cntxx1 = 0
    cntxx2 = 0
    cntxxx1 = []
    cntxxx2 = []
    old1 = 0
    old2 = 0
    for sentence in sentencesComments:
        ss = sid.polarity_scores(sentence)
        for type in sorted(ss):
            cnter += 1
            if (type == "neg" or type == "neu"):
                #sumNeg += ss[type]
                negY.append(ss[type])
                cntxx1 += 1
                cntxxx1.append(cntxx1)
                #cntxxx2.append(cntxx2)
                #posY.append(old2)
                #old1 = ss[type]
                #cntxx1 += 1
                cnter1 += 1
                cnterr1 += 1

            if (type == "pos"):
                #sumPos += ss[type]
                posY.append(ss[type])
                cntxx2+=1
                cntxxx2.append(cntxx2)
                #cntxxx1.append(cntxx1)
                #negY.append(old1)
                #old2 = ss[type]
                #cntxx2+=1
                cnter2 += 1
                cnterr2 += 1
    if (cnterr1 + cnterr2) != 0:
        pctneg = round(((cnterr1 / (cnterr1 + cnterr2)) * 100) , 2)
        pctpos = round(((cnterr2 / (cnterr1 + cnterr2)) * 100) , 2)
    if (cnterr1 + cnterr2) == 0:
        pctneg = 0
        pctpos = 0
    return posY, negY, cntxxx1, cntxxx2, pctneg, pctpos
```
